```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent Function
def gradient_descent(x, y):
    m_curr = b_curr = 0    # Initialize parameters
    rate = 0.01            # Learning rate
    n = len(x)             # Number of data points

    plt.scatter(x, y, color='red', marker='+', linewidth=5)  # Scatter plot of data points

    for i in range(10000):
        y_predicted = m_curr * x + b_curr  # Predicted y values

        # Compute the gradients
        md = -(2/n) * sum(x * (y - y_predicted))  # Derivative w.r.t m
        yd = -(2/n) * sum(y - y_predicted)  # Derivative w.r.t b

        # Update m and b
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd

        # Optionally plot the regression line after certain iterations
        if i % 100 == 0:
            plt.plot(x, y_predicted, color='green', linewidth=2)

    plt.show()


x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])


gradient_descent(x, y)
```
