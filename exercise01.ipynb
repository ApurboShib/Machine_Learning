```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df
```

```python
# my task is to predict the income the year of 2020.
```

```python
df  = df.rename(columns={'per capita income (US$)':'income'})
df
```

```python
# plot and draw the data.
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='red',marker='*')
```

```python
# create a linear regression object.

reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
# now predict the sample price of 2010.
prediction = reg.predict([[2020]])
prediction
```

```python
# check that my prediction is correct or not.

slope_m = reg.coef_
slope_m
```

```python
intercept_b = reg.intercept_
intercept_b
```

```python
# want to check the prediction of 'x = 2010'

x = 2020
```

```python
# calculate the answer.

func_of_y_is = slope_m * x + intercept_b
func_of_y_is
```
