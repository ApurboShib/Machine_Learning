# Sample problem of predicting home price in monroe, new jersey (USA).
# Problem Statement: Given above data build a machine learning model that can predict home prices based on square feet area

# You can represent values in above table as a scatter plot (values are shown in red markers). After that one can draw a straight line that best fits values on chart.

```python
# add necessary library needed to be import.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# firstly we need the dataframe.
# import a DataFrame(csv).

df = pd.read_csv("homeprices.csv")
df
```

```python
# secoundly we neet the plot of dataset.

# This is a magic command used to ensure that plots are displayed.
%matplotlib inline
plt.xlabel('Area(sqr ft)') # x-axis eh ki dite cacchi.
plt.ylabel('Price(USD)')  # y-axix eh ki dicchi.
plt.scatter(df.area,df.price,color='red',marker='+') # area/price er ekta plot.
```

```python
# create a linear regrassion object.
reg = linear_model.LinearRegression() # create an object.
reg.fit(df[['area']],df.price) # available data diye ekta plot draw korbo.(reg.fit()).
```

```python
# cheacking part..

# now check the sample predict price. using predict() function.
# here "3300" is a parameter.

prediction = reg.predict([[5000]])
prediction
```

```python
# see the co-efficient value which is called in my formula as 'm'.
slope_m = reg.coef_
slope_m
```

```python
# see the intecept value which is called in my formula as 'b'.

intercept_b = reg.intercept_
intercept_b
```

```python
# here 3300 is a predict value which is called as 'x'
```

```python
x = 5000
```

```python
# now calculate ..

# y = m * x + b

func_of_y_is =   slope_m * x + intercept_b
func_of_y_is
```

```python
# make the line that shows the regression line.

%matplotlib inline
plt.xlabel('Area', fontsize=20) # Set the label for the x-axis.
plt.ylabel('Price', fontsize=20) # Set the label for the y-axis.
plt.scatter(df.area,df.price,color='red',marker='+') # Create a scatter plot of the data points.
plt.plot(df.area,reg.predict(df[['area']]),color='blue') # Plot the regression line.
```
